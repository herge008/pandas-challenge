```python
import pandas as pd

# File to Load (Remember to Change These)
file = "C:/Users/Laptop/Desktop/GT-ATL-DATA-PT-12-2019-U-C/Homework/04-Pandas/Instructions/HeroesOfPymoli/Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file)
purchase_data.head()
```

```python
#total player count
#there are duplicate values....find unique!
SN_count = purchase_data['SN'].unique()
SN_count = len(SN_count)
data = {'Player Count':[SN_count]}
df = pd.DataFrame(data, index = ['0'])
df.head()
```

```python
#purchasing analysis total
ID_count = purchase_data['Item ID'].unique()
ID_count = len(ID_count)
Avg_Price = purchase_data['Price'].mean()
Ttl_Purch = purchase_data['Item Name'].count()
Ttl_Revenue = purchase_data['Price'].sum()
data_new = {'Number of Unique Items':[ID_count], 
            'Average Price': [Avg_Price], 
            'Number of Purchases': [Ttl_Purch], 
            'Total Revenue': Ttl_Revenue}
df_new = pd.DataFrame(data_new, index = ['0'])
df_new["Average Price"] = df_new["Average Price"].astype(float).map("${:,.2f}".format)
df_new["Total Revenue"] = df_new["Total Revenue"].astype(float).map("${:,.2f}".format)
df_new.head()
```

```python
#gender demographics
gender_data = pd.DataFrame(purchase_data.groupby(['Gender']).nunique())
gender_data = gender_data[["SN"]]
gender_data = gender_data.sort_values("SN", ascending=False)
gender_data = gender_data.rename(columns={"SN":"Total Count"})
gender_total = gender_data["Total Count"].sum()
gender_pct = (gender_data["Total Count"]/gender_total).astype(float).map('{:,.2%}'.format)
gender_data["Percentage of Players"] = gender_pct
gender_data
```

```python
#purchasing analysis (Gender)
gender_purch_data = pd.DataFrame(purchase_data.groupby(['Gender']).nunique())
gender_purch_data = gender_purch_data[["Purchase ID"]]
gender_purch_data = gender_purch_data.rename(columns={"Purchase ID":"Purchase Count"})
gender_purch_data['Purchase Count'] = pd.to_numeric(gender_purch_data['Purchase Count'])
gender_purch_data.head()
```

```python
gender_purch_data2 = pd.DataFrame(purchase_data.groupby(['Gender']).mean())
gender_purch_data2 = gender_purch_data2[["Price"]]
gender_purch_data2 = gender_purch_data2.rename(columns={"Price":"Average Purchase Price"})
gender_purch_data2['Average Purchase Price'] = pd.to_numeric(gender_purch_data2['Average Purchase Price'])
gender_purch_data2.head()
```

```python
gender_purch_data3 = pd.DataFrame(purchase_data.groupby(['Gender']).sum())
gender_purch_data3 = gender_purch_data3[["Price"]]
gender_purch_data3 = gender_purch_data3.rename(columns={"Price":"Total Purchase Value"})
gender_purch_data3['Total Purchase Value'] = pd.to_numeric(gender_purch_data3['Total Purchase Value'])
gender_purch_data3.head()
```

```python
gender_purch_data_merged = pd.merge(gender_purch_data, gender_purch_data2, on="Gender")
gender_purch_data_merged = pd.merge(gender_purch_data_merged, gender_purch_data3, on="Gender")
avg_per = gender_purch_data_merged["Total Purchase Value"] / gender_data["Total Count"]
gender_purch_data_merged["Avg Total Purchase per Person"] = avg_per
gender_purch_data_merged["Average Purchase Price"] = gender_purch_data_merged["Average Purchase Price"].astype(float).map("${:,.2f}".format)
gender_purch_data_merged["Total Purchase Value"] = gender_purch_data_merged["Total Purchase Value"].astype(float).map("${:,.2f}".format)
gender_purch_data_merged["Avg Total Purchase per Person"] = gender_purch_data_merged["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)
gender_purch_data_merged
```

```python
#Top Spenders
top_data = pd.DataFrame(purchase_data.groupby(['SN']).nunique())
top_data = top_data[["Item Name"]]
top_data = top_data.sort_values("Item Name", ascending=False)
top_data = top_data.rename(columns={"Item Name":"Purchase Count"})
top_data['Purchase Count'] = pd.to_numeric(top_data['Purchase Count'])
top_data.head()
```

```python
top_data2 = pd.DataFrame(purchase_data.groupby(['SN']).mean())
top_data2 = top_data2[["Price"]]
top_data2 = top_data2.rename(columns={"Price":"Average Purchase Price"})
top_data2['Average Purchase Price'] = pd.to_numeric(top_data2['Average Purchase Price'])
top_data2.head()
```

```python
top_data3 = pd.DataFrame(purchase_data.groupby(['SN']).sum())
top_data3 = top_data3[["Price"]]
top_data3 = top_data3.rename(columns={"Price":"Total Purchase Value"})
top_data3['Total Purchase Value'] = pd.to_numeric(top_data3['Total Purchase Value'])
top_data3.head()
```

```python
top_data_merged = pd.merge(top_data, top_data2, on="SN")
top_data_merged = pd.merge(top_data_merged, top_data3, on="SN")
top_data_merged["Average Purchase Price"] = top_data_merged["Average Purchase Price"].astype(float).map("${:,.2f}".format)
top_data_merged["Total Purchase Value"] = top_data_merged["Total Purchase Value"].astype(float).map("${:,.2f}".format)
top_data_merged.head()
```

```python
#Most Popular Items
popular_data = pd.DataFrame(purchase_data.groupby(['Item ID', 'Item Name']).nunique())
popular_data = popular_data[["Purchase ID"]]
popular_data = popular_data.sort_values("Purchase ID", ascending=False)
popular_data = popular_data.rename(columns={"Purchase ID":"Purchase Count"})
popular_data['Purchase Count'] = pd.to_numeric(popular_data['Purchase Count'])
popular_data.head()
```

```python
popular_data2 = pd.DataFrame(purchase_data.groupby(['Item ID', 'Item Name']).sum())
popular_data2 = popular_data2[["Price"]]
popular_data2 = popular_data2.rename(columns={"Price":"Total Purchase Value"})
popular_data2['Total Purchase Value'] = pd.to_numeric(popular_data2['Total Purchase Value'])
popular_data2.head()
```

```python
popular_data_merged = pd.merge(popular_data, popular_data2, on=["Item ID", "Item Name"])
item_price = popular_data_merged["Total Purchase Value"] / popular_data_merged["Purchase Count"]
popular_data_merged["Item Price"] = item_price
popular_data_merged["Total Purchase Value"] = popular_data_merged["Total Purchase Value"].astype(float).map("${:,.2f}".format)
popular_data_merged["Item Price"] = popular_data_merged["Item Price"].astype(float).map("${:,.2f}".format)
organized_popular_data_merged = popular_data_merged[["Purchase Count", "Item Price", "Total Purchase Value"]]
organized_popular_data_merged.head()
```

# Most Profitable Items

```python
profitable_data = pd.DataFrame(purchase_data.groupby(['Item ID', 'Item Name']).sum())
profitable_data = profitable_data[["Price"]]
profitable_data = profitable_data.sort_values("Price", ascending=False)
profitable_data = profitable_data.rename(columns={"Price":"Total Purchase Value"})
profitable_data_merged = pd.merge(profitable_data, organized_popular_data_merged, on=["Item ID", "Item Name"])
organized_profitable_data_merged = profitable_data_merged[["Purchase Count", "Item Price", "Total Purchase Value_y"]]
organized_profitable_data_merged = organized_profitable_data_merged.rename(columns={"Total Purchase Value_y":"Total Purchase Value"})
organized_profitable_data_merged.head()
```
